```python
# Remove the specified columns from the dataset
columns_to_remove = ['date', 'lights']
data = data.drop(columns_to_remove, axis=1)

# Define the feature matrix (X) and target variable (y)
X = data.drop('Appliances', axis=1)
y = data['Appliances']

# Normalize the feature matrix using MinMaxScaler
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Perform train-test split with a random state of 42
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=42)

# Fit the linear regression model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Print the MAE rounded to two decimal places
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
# Fit the linear regression model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate the Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_test, y_pred, squared=False)

# Calculate the Coefficient of Determination (R-squared)
r_squared = r2_score(y_test, y_pred)

# Print the RMSE rounded to three decimal places
print(f"Root Mean Squared Error: {rmse:.3f}")

# Print the R-squared value rounded to two decimal places
print(f"Coefficient of Determination (R-squared): {r_squared:.2f}")
```

```python
# Get the feature weights
feature_weights = pd.DataFrame({'Feature': X.columns, 'Weight': model.coef_})

# Sort the feature weights in ascending order
feature_weights = feature_weights.sort_values(by='Weight')

# Print the feature with the lowest weight
lowest_weight_feature = feature_weights.iloc[0]['Feature']
print(f"Feature with the lowest weight: {lowest_weight_feature}")

# Print the feature with the highest weight
highest_weight_feature = feature_weights.iloc[-1]['Feature']
print(f"Feature with the highest weight: {highest_weight_feature}")
```

```python
# Train a ridge regression model with alpha=0.4
ridge_model = Ridge(alpha=0.4)
ridge_model.fit(X_train, y_train)

# Predict on the test set using the ridge regression model
ridge_y_pred = ridge_model.predict(X_test)

# Calculate the RMSE for the ridge regression model
ridge_rmse = mean_squared_error(y_test, ridge_y_pred, squared=False)

# Train a lasso regression model with alpha=0.001
lasso_model = Lasso(alpha=0.001)
lasso_model.fit(X_train, y_train)

# Predict on the test set using the lasso regression model
lasso_y_pred = lasso_model.predict(X_test)

# Calculate the RMSE for the lasso regression model
lasso_rmse = mean_squared_error(y_test, lasso_y_pred, squared=False)

# Get the feature weights from the lasso regression model
lasso_feature_weights = pd.DataFrame({'Feature': X.columns, 'Weight': lasso_model.coef_})

# Count the number of features with non-zero weights
non_zero_features = len(lasso_feature_weights[lasso_feature_weights['Weight'] != 0])

# Print the changes in RMSE and the number of features with non-zero weights
print(f"Change in RMSE (Ridge vs. Linear Regression): {ridge_rmse:.3f}")
print(f"Number of features with non-zero weights (Lasso): {non_zero_features}")
print(f"RMSE with Lasso Regression: {lasso_rmse:.3f}")
```
